```python
import numpy as np
from matplotlib import pyplot as plt
import csv

"""
Sanjay Kumar, 000811237, The purposr of this script is to read the data from the caesarien.csv file and then 
divide the data in training set and testing set in 75% and 25%. Then plot three different scatter plots and and a bar graph.
"""


## Read it

filename = 'caesarian.csv'

featureName =['Age', 'Delivery Number','Delivery Time','Heart Pressure']
data = []
caesarien = []
with open(filename) as file:
    reader = csv.DictReader(file)
    for row in reader:
        data.append([int(row["Age"]),int(row["Delivery number"]),int(row["Delivery time"]), int(row["Blood of Pressure"]), int(row["Heart Problem"]),int(row["Caesarian"])])

featuresNames = np.array(featureName)
allData = np.array(data)
np.random.shuffle(allData)
features = allData[:,:-1]
labels = allData[:,-1]

splitPercentage = int(0.75*len(allData))

trainingData = allData[:splitPercentage]
testingData = allData[splitPercentage:]

trainingLabels = allData[:splitPercentage]
testingLabels = allData[splitPercentage:]

# Summarize it
print("TRAINING DATA")
print(featuresNames)
print("Maxima: ",trainingData.max(axis=0))
print("Minima: ",np.min(trainingData, axis = 0))
print("Mean: ", np.mean(trainingData, axis =0))
print("Median: ",np.median(trainingData,axis =0))
print()
print("TESTING DATA")
print(featuresNames)
print("Maxima: ",testingData.max(axis=0))
print("Minima: ",np.min(testingData, axis = 0))
print("Mean: ", np.mean(testingData, axis =0))
print("Median: ",np.median(testingData,axis =0))

## Graph it

plt.figure(1)
plt.scatter(trainingData[:,0],trainingData[:,1])
plt.xlabel("Ages")
plt.ylabel("Delivery Number")
plt.title("Ages vs. Delivery Number")
plt.show()

plt.figure(2)
plt.scatter(trainingData[:,0],trainingData[:,2])
plt.xlabel("Ages")
plt.ylabel("Delivery Time")
plt.title("Ages vs. Delivery Time")
plt.show()

plt.figure(3)
plt.scatter(trainingData[:,1],trainingData[:,4],c='y')
plt.xlabel("Delivery Number")
plt.ylabel("Heart of Pressure")
plt.title("Delivery Number vs Heart of pressure")
plt.show()

labeloccurance = np.bincount(labels)
plt.bar([0,1],[labeloccurance[0],labeloccurance[1]])
plt.xlabel("Values of labels")
plt.ylabel("Occurance of labels")
plt.title("Labels Frequency chart")
plt.show()
```
